# Fundamentals of Linear Algebra 

##  Vector Spaces and Subspaces
### Definition: Vector Space  

A vector space $V$ (or linear space) over a field $\mathbb{F}$ is a set with the following binary operations:
- Addition: $V\times V \rightarrow V$  
- Scalar Multiplication: $\mathbb{F}\times V \rightarrow V$
and satisfy the following axioms: $\forall \mathbf{u}, \mathbf{v}, \mathbf{w} \in V$ and $\forall \alpha \in \mathbb{F}$,
- Commutativity: $\mathbf{u} + \mathbf{v} = \mathbf{v} + \mathbf{u}$
- Associativity: $(\mathbf{u} + \mathbf{v}) + \mathbf{w} = \mathbf{u} + (\mathbf{v} + \mathbf{w})$
- Additive Identity: $\mathbf{u} + \mathbf{0} = \mathbf{u}$
- Additive Inverse: $\mathbf{u} + (-\mathbf{u}) = \mathbf{0}$
- Scalar Multiplication: $\alpha(\beta \mathbf{u}) = (\alpha \beta) \mathbf{u}$
- Identity element of scalar multiplication: $1 \mathbf{u} = \mathbf{u}$
- Scalar Assotiativity: $(\alpha + \beta) \mathbf{u} = \alpha \mathbf{u} + \beta \mathbf{u}$
- Scalar Distributivity: $\alpha(\mathbf{u} + \mathbf{v}) = \alpha \mathbf{u} + \alpha \mathbf{v}$

### Examples of Vector Spaces:
- The set of all $n$-tuples of real numbers $\mathbb{R}^n$.
- Space of polynomials of degree less than or equal to $n$.
- Space of continuous functions on a closed interval $[a, b]$.
- Space of differentiable functions on $[a, b]$.
- Space of square-integrable functions on $[a, b]$.
- Space of $m \times n$ matrices with real entries.

### Definition: Subspaces
A subset $W$ of a vector space $V$ is called a **subspace** of $V$ if it satisfies the following conditions:
- $W$ is non-empty.
- If $\mathbf{u}, \mathbf{v} \in W$, then $\mathbf{u} + \mathbf{v} \in W$.
- If $\mathbf{u} \in W$ and $\alpha \in \mathbb{R}$, then $\alpha \mathbf{u} \in W$.
On the other words, a subspace is a vector space in its own right, with the same operations as the original vector space.

### Examples of Subspaces:
- The set of all solutions to a homogeneous system of linear equations.
- The set of all polynomials of degree less than or equal to $n$.
- The set of all continuous functions on $[a, b]$ that are zero at $a$ and $b$.
- The set of all $m \times n$ matrices with real entries that have a certain property (e.g., symmetric, upper triangular).
- **Krylov subspace**: which is a sequence of vector spaces generated by repeatedly applying a matrix $A$ to a vector $b$. The $m$-th Krylov subspace is defined as:
$$
\mathcal{K}_m(A, b) = \text{span}\{b, Ab, A^2b, \ldots, A^{m-1}b\}
$$
Krylov subspaces are widely used in numerical linear algebra, especially for solving large systems of linear equations and eigenvalue problems (e.g., with the Arnoldi or Lanczos algorithms).



### Span and Linear independence
- **Span**: The span of a set of vectors $\{v_1, v_2, \ldots, v_n\}$ is the set of all linear combinations of those vectors.
$$\text{span}\{v_1, v_2, \ldots, v_n\} = \{a_1v_1 + a_2v_2 + \cdots + a_nv_n \mid a_1, a_2, \ldots, a_n \in \mathbb{R}\}$$
- **Linear independence**: A set of vectors is linearly independent if no vector in the set can be written as a linear combination of the other vectors in the set.
- **Basis**: A basis for a vector space is a linearly independent set of vectors that spans the space.
- **Dimension**: The dimension of a vector space is the number of vectors in any basis for the space.
